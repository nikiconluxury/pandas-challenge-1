## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
data = df
```

```python
data['subtotal'] = data['unit_price'] * data['qty']
selected_columns = data[['unit_price', 'qty', 'subtotal']]
selected_columns.head(2)
```

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

data['subtotal'] = data['unit_price'] * data['qty']
selected_columns = data[['unit_price', 'qty', 'subtotal']]
selected_columns.head(2)
```

```python
data['t_weight'] = data['unit_weight'] * data['qty']
```

```python
# Step 2: Define the shipping price calculation
def calculate_shipping_price(row):
    if row['t_weight'] > 50:
        return 7 * row['t_weight']
    else:
        return 10 * row['t_weight']
```

```python
# Apply the shipping price calculation
data['shipping_price'] = data.apply(calculate_shipping_price, axis=1)
data[['unit_price', 'unit_weight', 'qty', 't_weight', 'shipping_price']].head(3)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

```

```python

data['total_price'] = (data['subtotal'] + data['shipping_price']) * (1 + 0.0925)
data[['subtotal', 'shipping_price', 'total_price']].head(3)

```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%


data['total_price'] = (data['subtotal'] + data['shipping_price']) * (1 + 0.0925)
data[['subtotal', 'shipping_price', 'total_price']].head(3)

```

```python

data['line_cost'] = (data['unit_cost'] * data['qty']) + data['shipping_price']
data.head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

data['line_cost'] = (data['unit_cost'] * data['qty']) + data['shipping_price']
data.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price
data['profit'] = (data['total_price'] - data['line_cost'])
data.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

data['profit'] = (data['total_price'] - data['line_cost'])
data.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
summary={}
for client in top_clients:
    total= data[data['client_id'] == client]['total_price'].sum()
    summary[client] = total
summary  
```

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
summary={}
for client in top_clients:
    total= data[data['client_id'] == client]['total_price'].sum()
    summary[client] = total
summary  
```

```python
top_client_ids = [24741, 38378, 66037, 46820, 33615]
top_clients_data = data[data['client_id'].isin(top_client_ids)]
top_clients_data
```

```python
summary = top_clients_data.groupby('client_id').agg(
    total_units_purchased=pd.NamedAgg(column='qty', aggfunc='sum'),
    total_shipping_price=pd.NamedAgg(column='shipping_price', aggfunc='sum'),
    total_revenue=pd.NamedAgg(column='total_price', aggfunc='sum'),
    total_cost=pd.NamedAgg(column='line_cost', aggfunc='sum')
)
summary['total_profit'] = summary['total_revenue'] - summary['total_cost']
```

```python
summary_sorted_by_profit = summary.sort_values(by='total_profit', ascending=False)
```

```python
data_summary = summary_sorted_by_profit.reset_index()
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

data_summary
```

```python

def format_currency_millions(value):
    return f"${value / 1e6:.2f}M"

columns_to_format = ['total_units_purchased', 'total_shipping_price', 'total_revenue', 'total_cost', 'total_profit']
for column in columns_to_format:
    summary_sorted[column] = summary_sorted[column].apply(format_currency_millions)
```

```python
summary_sorted = data_summary.sort_values('total_profit', ascending=False)
```

```python
summary_sorted.rename(columns={
    'total_units_purchased': 'Units',
    'total_shipping_price': 'Shipping',
    'total_revenue': 'Total Revenue',
    'total_cost': 'Total Cost',
    'total_profit': 'Total Profit'
}, inplace=True)
```
